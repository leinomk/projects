## Datan lukeminen Jupyter-työkirjassa


```python
import pandas as pd
import numpy as np
import os
```

```python
#Ladataan AirBNB-data, ja valitaan todennäköisesti kiinnostavimmat kolumnit datasetistä

LISTINGSFILE = 'http://data.insideairbnb.com/united-kingdom/england/london/2021-12-07/data/listings.csv.gz'

cols = ['id',
        'price',
        'review_scores_rating',
        'neighbourhood_cleansed',
        'latitude',
        'longitude',
        'property_type',
        'room_type',
        'accommodates',
        'amenities',
        'bedrooms',
        'beds',
        'bathrooms_text',
        'reviews_per_month',
        'number_of_reviews',
        'cleaning_fee'
        ]

if os.path.exists('listings.csv'):
    df = pd.read_csv('listings.csv', usecols=cols, low_memory=False)
else:
    df = pd.read_csv(LISTINGSFILE, usecols=cols, low_memory=False, compression='gzip')
    df.to_csv('listings.csv')

print(df.info())
```

Katsotaan vielä, millaisia arvoja 'bathrooms_text'-kolumnissa on ja kuinka paljon:

```python
df["bathrooms_text"].value_counts()
```

Nämä arvot eivät välttämättä ole kovin tarpeellisia tai kiinnostavia, mutta niitä pystyy mahdollisesti siivoamaan käyttökelpoisemmaksi, joten pidetään ne vielä mukana. Tehdään vastaavanlainen lataus myös pubeista tarjolla olevalle datasetille:

```python
PUBSFILE = 'https://data.london.gov.uk/download/cultural-infrastructure-map/7e6b0f36-da01-4029-99d7-640d89c4a0a4/Pubs.csv'

if os.path.exists('pubs.csv'):
    df = pd.read_csv('pubs.csv')
else:
    df = pd.read_csv(PUBSFILE)
    df.to_csv('pubs.csv')

print(df.info())
```

```python
df.head()
```

Näistä kolumneista tarvitaan todennäköisesti oikeastaan ainoastaan pubin sijainnin leveys- ja pituuspiiri, jotta pubin sijainti yksittäisestä AirBNB-asunnosta voidaan laskea. Vastaavalla tyylillä voidaan ladata vielä loputkin datasetit. Datan muokkaamisesta ja datasettien yhdistämisestä lisää seuraavassa vaiheessa.

```python
MUSEUMSFILE = 'https://data.london.gov.uk/download/cultural-infrastructure-map/d091e270-182a-444d-b4ee-64d9d9682749/Museums_and_public_galleries.csv'
MUSICVENUESFILE = 'https://data.london.gov.uk/download/cultural-infrastructure-map/5effe6f3-3697-49c4-8dcb-d9db58e7e00b/Music_venues_grassroots.csv'
CINEMASFILE = 'https://data.london.gov.uk/download/cultural-infrastructure-map/f353f97b-a1fb-41fa-810a-825586839b4d/Cinemas.csv'

if os.path.exists('museums.csv'):
    df = pd.read_csv('museums.csv')
else:
    df = pd.read_csv(MUSEUMSFILE)
    df.to_csv('museums.csv')
print(df.info())
    
if os.path.exists('musicvenues.csv'):
    df = pd.read_csv('musicvenues.csv')
else:
    df = pd.read_csv(MUSICVENUESFILE)
    df.to_csv('musicvenues.csv')
print(df.info())

if os.path.exists('cinemas.csv'):
    df = pd.read_csv('cinemas.csv')
else:
    df = pd.read_csv(CINEMASFILE)
    df.to_csv('cinemas.csv')
print(df.info())
```
